## BTC Only (Baseline)

```python
BTC_ONLY_PORTFOLIO = 'Buy & Hold BTC'
buy_hold_btc_weights = pd.Series(np.zeros(len(ASSETS)), index=ASSETS)
buy_hold_btc_weights.loc['BTC'] = 1.0
experiments[BTC_ONLY_PORTFOLIO] = buy_hold_btc_weights
buy_hold_btc_weights
```

```python
experiments[BTC_ONLY_PORTFOLIO].plot.pie(
                                       figsize=(8, 8),
                                       autopct='%1.1f%%',
                                       label='Portfolio Allocation')
```

## Balance weight

```python
BALANCE_WEIGHT_PORTFOLIO = 'Balance weight'
balance_weight_weights = pd.Series(np.full(len(ASSETS), 1 / len(ASSETS)), index=ASSETS)

experiments[BALANCE_WEIGHT_PORTFOLIO] = balance_weight_weights
experiments[BALANCE_WEIGHT_PORTFOLIO]
```

```python
experiments[BALANCE_WEIGHT_PORTFOLIO].plot.pie(
                                       figsize=(8, 8),
                                       autopct='%1.1f%%',
                                       label='Portfolio Allocation')
```

## Market-implied return Allocation

```python
MARKET_IMPLIED_RETURN_PORTFOLIO = 'Market-implied return + Maximize Sharpe ratio'
```

### Use BTC as market index

```python
market_index_prices = price_df.mean(axis=1)
delta = black_litterman.market_implied_risk_aversion(market_index_prices, frequency=len(price_df))
prior = black_litterman.market_implied_prior_returns(market_caps, delta, corr_df)
```

```python
delta
```

```python
prior
```

### Maximize Sharpe ratio 
penalty with L2 Regularization

```python
ef = EfficientFrontier(prior, corr_df.values)
ef.add_objective(objective_functions.L2_reg)
ef.max_sharpe()

experiments[MARKET_IMPLIED_RETURN_PORTFOLIO] = pd.Series(ef.clean_weights())
experiments[MARKET_IMPLIED_RETURN_PORTFOLIO]
```

```python
experiments[MARKET_IMPLIED_RETURN_PORTFOLIO].plot.pie(
                                               figsize=(8, 8),
                                               autopct='%1.1f%%',
                                               label='Portfolio Allocation')
```

# Visualization


```python
class PortfolioBacktester:
    def __init__(self, price_df, portfolio_weight_df):
        self.price_df = price_df
        self.portfolio_weight_df = portfolio_weight_df
        
        self.calculate_portfolio_value()
    def calculate_portfolio_value(self):
        init_price = self.price_df.iloc[0]
        self.position_size_df = self.portfolio_weight_df.div(init_price, axis=0)

        self.portfolio_value_df = self.position_size_df.apply(lambda weights: (self.price_df*weights).sum(axis=1))
        
    def plot_value(self, name=None, **kwargs):
        self.portfolio_value_df.plot(**kwargs)
    
    def plot_value_area(self, portfolio_name=None, **kwargs):
        if portfolio_name is not None:
            asset_value_df = self.price_df * self.position_size_df[portfolio_name]
            asset_value_df.plot.area(title=portfolio_name, **kwargs)
        else:
            fig, axes = plt.subplots(len(self.portfolio_weight_df.columns))
            for ax, portfolio_name in zip(axes, self.portfolio_weight_df.columns):
                asset_value_df = self.price_df * self.position_size_df[portfolio_name]
                asset_value_df.plot.area(title=portfolio_name, ax=ax, **kwargs)

portfolio_backtester = PortfolioBacktester(price_df, experiments)
```

```python
portfolio_backtester.plot_value(figsize=(16, 6))
```

```python
portfolio_backtester.plot_value_area(figsize=(16, 18))
```
